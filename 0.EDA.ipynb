```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import rcParams

#数据读取与预处理
df = pd.read_csv('../数据/Dataset for diabetes research.csv', encoding='gbk', na_values=['', ' ', 'NA', 'NULL', 'null'])
df.columns = [col.lower() for col in df.columns]
#核心特征
features = ['ggt', 'tp', 'ast', 'age', 'hr', 'alt', 'mono', 'umalb']

df_clean = df[features].copy()

rcParams['font.family'] = 'Arial'
rcParams['font.size'] = 10
rcParams['axes.linewidth'] = 1.0
rcParams['pdf.fonttype'] = 42
rcParams['ps.fonttype'] = 42
plt.rcParams['xtick.direction'] = 'out'
plt.rcParams['ytick.direction'] = 'out'
#异常值计算函数
def count_outliers(data):
    Q1 = data.quantile(0.25)
    Q3 = data.quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    return ((data < lower) | (data > upper)).sum()
n_cols = 4
n_rows = 2
fig, axes = plt.subplots(n_rows, n_cols, figsize=(14, 7))
axes = axes.flatten()
for idx, col in enumerate(features):
    ax = axes[idx]
    data = df_clean[col].dropna()
    n = len(data)
    outlier_num = count_outliers(data)
    box = ax.boxplot(
        data,
        widths=0.6,
        patch_artist=True,
        showfliers=True,
        medianprops={'color': '#000000', 'linewidth': 1.5},
        whiskerprops={'color': '#000000', 'linewidth': 1.0},
        capprops={'color': '#000000', 'linewidth': 1.0},
        flierprops={'marker': 'o', 'markerfacecolor': '#D55E00', 'markersize': 3, 'alpha': 0.8}
    )
    box['boxes'][0].set_facecolor('white')
    box['boxes'][0].set_edgecolor('black')
    box['boxes'][0].set_linewidth(1.0)

    ax.set_title(f'{col.upper()}\nN={n} | Outliers={outlier_num}', fontsize=10, pad=8)
    ax.set_ylabel(col.upper(), fontsize=9)
    ax.set_xticks([])
    ax.set_xlim(0.2, 1.8)
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)

for j in range(idx + 1, len(axes)):
    axes[j].axis('off')

plt.tight_layout(pad=2.0)
plt.savefig('箱线图.png', dpi=600, bbox_inches='tight')
plt.show()
```
