```python
from sklearn.neural_network import MLPClassifier
import pandas as pd
from sklearn.model_selection import cross_validate, StratifiedKFold

from imblearn.pipeline import Pipeline as ImbPipeline

import xgboost as xgb
import lightgbm as lgb

from sklearn.ensemble import (
    RandomForestClassifier, ExtraTreesClassifier, GradientBoostingClassifier, BaggingClassifier, HistGradientBoostingClassifier
)
from sklearn.tree import DecisionTreeClassifier

from imblearn.combine import SMOTEENN
from sklearn.neighbors import KNeighborsClassifier

df = pd.read_csv('../数据/训练集.csv')

print("数据形状:", df.shape)
print("目标变量分布:\n", df['ms_cds'].value_counts())

tezheng = ['ggt', 'drinking', 'ast', 'age', 'hr', 'alt', 'mono', 'gender', 'umalb', 'sua', 'dbil', 'bun', 'tp', 'idbil', 'rbc']
X = df[tezheng]
y = df['ms_cds']

models = {
    "SHINEMS": ImbPipeline([
        ('sampler', SMOTEENN(random_state=42, n_jobs=-1)),
        ('classifier', xgb.XGBClassifier(n_jobs=-1, random_state=42, eval_metric='logloss', use_label_encoder=False))
    ]),
    "LightGBM": ImbPipeline([
        ('sampler', SMOTEENN(random_state=42, n_jobs=-1)),
        ('classifier', lgb.LGBMClassifier(n_jobs=-1, random_state=42, verbose=-1))
    ]),


    "Hist Gradient Boosting": ImbPipeline([
        ('sampler', SMOTEENN(random_state=42, n_jobs=-1)),
        ('classifier', HistGradientBoostingClassifier(random_state=42))
    ]),
    "Random Forest": ImbPipeline([
        ('sampler', SMOTEENN(random_state=42, n_jobs=-1)),
        ('classifier', RandomForestClassifier(n_jobs=-1, random_state=42))
    ]),
    "Extra Trees": ImbPipeline([
        ('sampler', SMOTEENN(random_state=42, n_jobs=-1)),
        ('classifier', ExtraTreesClassifier(n_jobs=-1, random_state=42))
    ]),
    "Decision Tree": ImbPipeline([
        ('sampler', SMOTEENN(random_state=42, n_jobs=-1)),
        ('classifier', DecisionTreeClassifier(random_state=42))
    ]),
    "Bagging": ImbPipeline([
        ('sampler', SMOTEENN(random_state=42, n_jobs=-1)),
        ('classifier', BaggingClassifier(n_jobs=-1, random_state=42))
    ]),
    "Gradient Boosting": ImbPipeline([
        ('sampler', SMOTEENN(random_state=42, n_jobs=-1)),
        ('classifier', GradientBoostingClassifier(random_state=42))
    ]),

    "KNN": ImbPipeline([
    ('sampler', SMOTEENN(random_state=42, n_jobs=-1)),
    ('classifier', KNeighborsClassifier(n_jobs=-1))
    ]),
    "MLP": ImbPipeline([
        ('sampler', SMOTEENN(random_state=42, n_jobs=-1)),
        ('classifier', MLPClassifier(random_state=42, max_iter=500))
    ])
}

cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

scoring = {
    'accuracy': 'accuracy',
    'precision': 'precision',
    'recall': 'recall',
    'f1': 'f1',
    'roc_auc': 'roc_auc'
}

results = []
for name, model in models.items():
    scores = cross_validate(model, X, y, cv=cv, scoring=scoring, n_jobs=-1)

    avg_accuracy = scores['test_accuracy'].mean()
    avg_precision = scores['test_precision'].mean()
    avg_recall = scores['test_recall'].mean()
    avg_f1 = scores['test_f1'].mean()
    avg_auc = scores['test_roc_auc'].mean()

    results.append({
        'Model': name,
        'Accuracy': avg_accuracy,
        'Precision': avg_precision,
        'Recall': avg_recall,
        'F1-Score': avg_f1,
        'AUC': avg_auc
    })

    print(f"\n=== {name} ===")
    print(f"平均准确率: {avg_accuracy:.4f}")
    print(f"平均精确率: {avg_precision:.4f}")
    print(f"平均召回率: {avg_recall:.4f}")
    print(f"平均F1-Score: {avg_f1:.4f}")
    print(f"平均AUC值: {avg_auc:.4f}")

results_df = pd.DataFrame(results)
results_df.to_excel('模型性能.xlsx', index=False)
print(results_df.to_string(index=False))
```

```python
import pandas as pd

import seaborn as sns
from sklearn.model_selection import cross_validate, StratifiedKFold
from imblearn.pipeline import Pipeline as ImbPipeline
from imblearn.combine import SMOTEENN

import xgboost as xgb
import lightgbm as lgb
from sklearn.ensemble import (
    ExtraTreesClassifier, BaggingClassifier, HistGradientBoostingClassifier
)
from sklearn.tree import DecisionTreeClassifier

import matplotlib.pyplot as plt
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False
df = pd.read_csv('../数据/训练集.csv')
tezheng = ['ggt', 'drinking', 'ast', 'age', 'hr', 'alt', 'mono', 'gender', 'umalb', 'sua', 'dbil', 'bun', 'tp', 'idbil', 'rbc']
X = df[tezheng]
y = df['ms_cds']


models = {
    "SHINEMS": ImbPipeline([
        ('sampler', SMOTEENN(random_state=42, n_jobs=-1)),
        ('classifier', xgb.XGBClassifier(n_jobs=-1, random_state=42, eval_metric='logloss', use_label_encoder=False))
    ]),
    "LightGBM": ImbPipeline([
        ('sampler', SMOTEENN(random_state=42, n_jobs=-1)),
        ('classifier', lgb.LGBMClassifier(n_jobs=-1, random_state=42, verbose=-1))
    ]),
    "Hist Gradient Boosting": ImbPipeline([
        ('sampler', SMOTEENN(random_state=42, n_jobs=-1)),
        ('classifier', HistGradientBoostingClassifier(random_state=42))
    ]),
    "Extra Trees": ImbPipeline([
        ('sampler', SMOTEENN(random_state=42, n_jobs=-1)),
        ('classifier', ExtraTreesClassifier(n_jobs=-1, random_state=42))
    ]),
    "Decision Tree": ImbPipeline([
        ('sampler', SMOTEENN(random_state=42, n_jobs=-1)),
        ('classifier', DecisionTreeClassifier(random_state=42))
    ]),
    "Bagging": ImbPipeline([
        ('sampler', SMOTEENN(random_state=42, n_jobs=-1)),
        ('classifier', BaggingClassifier(n_jobs=-1, random_state=42))
    ])
}

cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
scoring = {'roc_auc': 'roc_auc'}

fold_scores = {}
mean_scores = {}

for name, model in models.items():
    scores = cross_validate(model, X, y, cv=cv, scoring=scoring, n_jobs=-1)
    auc_folds = scores['test_roc_auc']
    fold_scores[name] = auc_folds
    mean_scores[name] = auc_folds.mean()
    print(f"{name}: AUC = {mean_scores[name]:.4f} ± {auc_folds.std():.4f}")

# 按 AUC 均值降序排列模型
sorted_models = sorted(mean_scores.keys(), key=lambda x: mean_scores[x], reverse=True)

plot_data = []
for model in sorted_models:
    for auc_val in fold_scores[model]:
        plot_data.append({'Model': model, 'AUC': auc_val})
plot_df = pd.DataFrame(plot_data)

plt.rcParams.update({
    'font.family': 'Arial',
    'font.size': 12,
    'axes.labelsize': 14,
    'axes.titlesize': 16,
    'xtick.labelsize': 10,
    'ytick.labelsize': 12,
    'legend.fontsize': 10,
    'figure.dpi': 300,
    'savefig.dpi': 300,
    'savefig.format': 'pdf',
    'savefig.bbox': 'tight',
    'axes.linewidth': 1.2,
})

fig, ax = plt.subplots(figsize=(14, 7))

boxprops = dict(linewidth=1.5, edgecolor='black')  # 黑色箱体边框
medianprops = dict(linewidth=1.5, color='#ff7f0e')  # 亮橙色中位数线
whiskerprops = dict(linewidth=1.2, color='black')  # 黑色须线
capprops = dict(linewidth=1.2, color='black')      # 黑色帽线
flierprops = dict(
    marker='*',
    markerfacecolor='red',
    markeredgecolor='red',
    markersize=6,
    linestyle='none'
)

sns.boxplot(
    data=plot_df, x='Model', y='AUC', order=sorted_models,
    palette=['#1f77b4'] * len(sorted_models),
    width=0.6,
    boxprops=boxprops,
    medianprops=medianprops,
    whiskerprops=whiskerprops,
    capprops=capprops,
    flierprops=flierprops,
    ax=ax
)

ax.set_title('5-Fold Cross-Validation AUC for All Models', fontweight='bold', pad=15)
ax.set_xlabel('')
ax.set_ylabel('AUC', fontweight='bold')
ax.set_ylim(0.55, 0.85)  # 根据实际数据范围微调
ax.grid(axis='y', linestyle='-', alpha=0.3, color='gray')
ax.set_xticklabels(ax.get_xticklabels(), rotation=45, ha='right')

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

plt.tight_layout()


plt.savefig('AUC_箱型图.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import cross_validate, StratifiedKFold
from sklearn.metrics import roc_curve, auc
from imblearn.pipeline import Pipeline as ImbPipeline
from imblearn.combine import SMOTEENN

# 模型库
import lightgbm as lgb
from sklearn.ensemble import BaggingClassifier
from sklearn.tree import DecisionTreeClassifier

df = pd.read_csv('../数据/训练集.csv')
tezheng = ['ggt', 'drinking', 'ast', 'age', 'hr', 'alt', 'mono', 'gender', 'umalb', 'sua', 'dbil', 'bun', 'tp', 'idbil', 'rbc']
X = df[tezheng]
y = df['ms_cds']

models = {
    "SHINEMS": ImbPipeline([
        ('sampler', SMOTEENN(random_state=42, n_jobs=-1)),
        ('classifier', xgb.XGBClassifier(n_jobs=-1, random_state=42, eval_metric='logloss', use_label_encoder=False))
    ]),
    "LightGBM": ImbPipeline([
        ('sampler', SMOTEENN(random_state=42, n_jobs=-1)),
        ('classifier', lgb.LGBMClassifier(n_jobs=-1, random_state=42, verbose=-1))
    ]),
    "Bagging": ImbPipeline([
        ('sampler', SMOTEENN(random_state=42, n_jobs=-1)),
        ('classifier', BaggingClassifier(n_jobs=-1, random_state=42))
    ]),
    "Decision Tree": ImbPipeline([
        ('sampler', SMOTEENN(random_state=42, n_jobs=-1)),
        ('classifier', DecisionTreeClassifier(random_state=42))
    ])
}

cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
scoring = {'roc_auc': 'roc_auc'}

fold_scores = {}
mean_scores = {}

for name, model in models.items():
    scores = cross_validate(model, X, y, cv=cv, scoring=scoring, n_jobs=-1)
    auc_folds = scores['test_roc_auc']
    fold_scores[name] = auc_folds
    mean_scores[name] = auc_folds.mean()
    print(f"{name}: AUC = {mean_scores[name]:.4f} ± {auc_folds.std():.4f}")

sorted_models = sorted(mean_scores.keys(), key=lambda x: mean_scores[x], reverse=True)

style_dict = {
    'SHINEMS': {'color': 'red', 'lw': 2.5, 'linestyle': '-'},
    'LightGBM': {'color': 'blue', 'lw': 2, 'linestyle': '--'},
    'Bagging': {'color': 'green', 'lw': 2, 'linestyle': '--'},
    'Decision Tree': {'color': 'orange', 'lw': 2, 'linestyle': '--'}
}

mean_fpr = np.linspace(0, 1, 1000)
roc_results = {}

for name in sorted_models:
    model = models[name]
    tpr_list = []
    auc_list = []

    for train_idx, test_idx in cv.split(X, y):
        X_train, X_test = X.iloc[train_idx], X.iloc[test_idx]
        y_train, y_test = y.iloc[train_idx], y.iloc[test_idx]

        model.fit(X_train, y_train)
        y_proba = model.predict_proba(X_test)[:, 1]

        fpr, tpr, _ = roc_curve(y_test, y_proba)
        roc_auc = auc(fpr, tpr)
        auc_list.append(roc_auc)

        tpr_interp = np.interp(mean_fpr, fpr, tpr)
        tpr_interp[0] = 0.0
        tpr_list.append(tpr_interp)

    mean_tpr = np.mean(tpr_list, axis=0)
    mean_auc = np.mean(auc_list)
    std_auc = np.std(auc_list, ddof=1)
    n_folds = len(auc_list)
    ci = 1.96 * std_auc / np.sqrt(n_folds)   # 95% 置信区间半宽

    roc_results[name] = {
        'mean_fpr': mean_fpr,
        'mean_tpr': mean_tpr,
        'mean_auc': mean_auc,
        'ci': ci
    }

# 绘图
fig, ax = plt.subplots(figsize=(10, 8))

for name in sorted_models:
    res = roc_results[name]
    style = style_dict[name]
    ax.plot(res['mean_fpr'], res['mean_tpr'],
            color=style['color'], lw=style['lw'], linestyle=style['linestyle'],
            label=f"{name} (AUC = {res['mean_auc']:.3f} ± {res['ci']:.3f})")

ax.plot([0, 1], [0, 1], color='gray', lw=1.5, linestyle='--', alpha=0.7)
ax.text(0.55, 0.52, 'Random guess (AUC = 0.5)',
        fontsize=10, color='gray', rotation=42, alpha=0.8)

ax.set_xlim(0, 1)
ax.set_ylim(0, 1.02)
ax.set_xlabel('False Positive Rate (1 - Specificity)', fontweight='bold', fontsize=14)
ax.set_ylabel('True Positive Rate (Sensitivity)', fontweight='bold', fontsize=14)
ax.set_title('Comparison of Receiver Operating Characteristic Curves',
             fontweight='bold', fontsize=16, pad=15)

# 网格与边框
ax.grid(axis='both', linestyle='-', alpha=0.2, color='gray')
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

# 图例（按 AUC 降序排列）
ax.legend(loc='lower right', frameon=True, fancybox=False, shadow=False,
          fontsize=11, edgecolor='k', facecolor='white', framealpha=0.9)

plt.tight_layout()
plt.savefig('ROC曲线图.png', dpi=600, bbox_inches='tight')
plt.show()
```
